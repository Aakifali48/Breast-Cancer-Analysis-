```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
X = np.array([74.679,74.8008,74.8534,74.9977,75.0076,75.3844,75.8201,75.6413,75.7903,75.5671,75.9632,76.2606,76.2606,76.2541,76.0273,76.1179,76.015,75.8179,75.8957,76.0183,76.2416,76.1352,76.3673,76.1759,76.3407,76.3407,76.21,76.4895,76.3061,76.5621]).reshape(30,1)
y = np.array([1344.75,1347.75,1361,1349.25,1354.75,1324,1306.75,1320.5,1307.25,1319,1307.25,1285.5,1282.5,1281.25,1272.5,1286,1287.25,1283.5,1275.75,1257,1240,1246.25,1243,1247.5,1245,1245.5,1253,1229.5,1217.25,1227.5]).reshape(30,1)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
MS = MinMaxScaler()
X_data = MS.fit_transform(X)
y_data = MS.fit_transform(y)
```

```python
def LR_fit(X, y, slope, yInt, alpha, iteration):
    n = len(X_data)
    for i in range(iteration):
        y_pred = slope*X+yInt
        print('Y-Predicted: ', str(y_pred))
        SE = 1/2 * np.square(y-y_pred)
        print('Squared Error: ', str(SE))
        SSE = np.sum(SE)
        print('Sum of SE: ', str(SSE))
        print('________________________________________')
# -------------------------------------------------------------------------------------------
        yInt_grad = np.sum(-(y-y_pred))
        print('Y-gradient: ', str(yInt_grad))
        slope_grad = np.sum(-(X*(y- y_pred)))
        print('Slope gradient: ', str(slope_grad))
        print('________________________________________')
        yInt = yInt - (alpha * yInt_grad)
        slope = slope - (alpha * slope_grad)
        print ('New Y-Intercept: ', str(yInt))
        print ('New slope: ', str(slope))
        print('________________________________________')
    return yInt, slope
```

```python
yInt, slope = LR_fit(X_data, y_data, 0.75, 0.45, 0.01, 10)
```

```python
plt.scatter(X_data,y_data)
plt.plot(X_data, X_data*slope+yInt)
plt.show()
```
